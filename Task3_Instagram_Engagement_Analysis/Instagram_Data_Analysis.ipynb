```python
import zipfile

with zipfile.ZipFile('/content/archive (3).zip', 'r') as zip_ref:
    zip_ref.extractall('/content')

print("ZIP extracted successfully!")
```

```python
import os

print(os.listdir('/content'))
```

```python
import pandas as pd

users = pd.read_csv('/content/users.csv')
photos = pd.read_csv('/content/photos.csv')
likes = pd.read_csv('/content/likes.csv')
comments = pd.read_csv('/content/comments.csv')
tags = pd.read_csv('/content/tags.csv')
photo_tags = pd.read_csv('/content/photo_tags.csv')
follows = pd.read_csv('/content/follows.csv')
```

```python
print("Users:", users.shape)
print("Photos:", photos.shape)
print("Likes:", likes.shape)
print("Comments:", comments.shape)
print("Tags:", tags.shape)
print("Photo Tags:", photo_tags.shape)
print("Follows:", follows.shape)
```

```python
print(users.isnull().sum())
print(photos.isnull().sum())
print(likes.isnull().sum())
print(comments.isnull().sum())
print(tags.isnull().sum())
print(photo_tags.isnull().sum())
print(follows.isnull().sum())
```

```python
print(photos.columns)
```

```python
import matplotlib.pyplot as plt

top_users = photos['user ID'].value_counts().head(10)

plt.figure(figsize=(8,5))
top_users.plot(kind='bar')
plt.title('Top 10 Most Active Users')
plt.xlabel('User ID')
plt.ylabel('Number of Posts')
plt.show()
```

```python
top_tags = tags['tag text'].value_counts().head(10)

plt.figure(figsize=(8,5))
top_tags.plot(kind='bar', color='orange')
plt.title('Top 10 Hashtags')
plt.xlabel('Hashtags')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
top_likes = likes['photo'].value_counts().head(10)

plt.figure(figsize=(8,5))
top_likes.plot(kind='bar', color='green')
plt.title('Top 10 Most Liked Photos')
plt.xlabel('Photo ID')
plt.ylabel('Likes')
plt.show()
```

```python
top_comments = comments['Photo id'].value_counts().head(10)

plt.figure(figsize=(8,5))
top_comments.plot(kind='bar', color='red')
plt.title('Top 10 Most Commented Photos')
plt.xlabel('Photo ID')
plt.ylabel('Comments')
plt.show()
```

```python
print(follows.columns)
```

```python
follows.columns = follows.columns.str.strip()
print(follows.columns)
```

```python
top_followed = follows['followee'].value_counts().head(10)

plt.figure(figsize=(8,5))
top_followed.plot(kind='bar', color='purple')
plt.title('Top 10 Most Followed Users')
plt.xlabel('User ID')
plt.ylabel('Followers')
plt.show()
```

```python
top_active = photos['user ID'].value_counts().head(10)

plt.figure(figsize=(8,5))
top_active.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Active Users')
plt.xlabel('User ID')
plt.ylabel('Number of Posts')
plt.show()
```
